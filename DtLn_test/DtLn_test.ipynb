```python
import pandas as pd
import numpy as np

# Models and metrics
from sklearn import tree 
from sklearn import metrics 

# Data optimization
from sklearn.preprocessing  import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
import category_encoders as ce

# Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly
import plotly.express as px  
import plotly.figure_factory as ff  
import plotly.io as pio
pio.renderers.default = "svg"

pd.set_option('display.max_columns', None)
```

```python
train_df = pd.read_csv("Data/train.csv")
test_df = pd.read_csv("Data/test.csv")
players_df = pd.read_csv("Data//players_feats.csv")
```

```python
# Create list of player's features
players_cols = players_df.drop(["map_name", "map_id", "team_id"],axis=1).columns.to_list()
pl_id_list = ["p1_id", "p2_id", "p3_id", "p4_id", "p5_id"]
for i in pl_id_list:
    players_cols.remove(i)

# Create list of unique features for each player
list_of_unique_feats = list()
for i in players_cols:
    if "p1_" in i:
        list_of_unique_feats.append(i.replace("p1_", ""))

def team_feats(df, list_of_cols, unique_feats):

    """ Function creates team features from players' features 
        Input:
                df[DataFrame]: dataframe with players' features
                list_of_cols[list]: list with names of features in df
                unique_feats[list]: list with unique features for each player
                pl_ids[list]: list with names of players' id columns
        Return:
                DataFrame with team features 
    """

    for i in unique_feats:
        # Create temporary list to add specific players' feature
        features = list()
        # Choose specific feature from all players
        for j in list_of_cols:
            if i in j:
                features.append(j)
                # Create specific feature with mean value for each team 
                df[i] = df[features].sum(axis=1)/5

    # Drop unnecessary columns 
    df = df.drop(list_of_cols, axis=1)
    return df


players_df = team_feats(players_df, players_cols, list_of_unique_feats)
players_df = players_df.drop(pl_id_list, axis=1)

# Merge team features in one dataset
# Create index lists for team1 and team2
lst_tm1 = list(range(0, 1485, 2))
lst_tm2 = list(range(1, 1486, 2))

team1_df = players_df.loc[lst_tm1]
team2_df = players_df.loc[lst_tm2]

team_df = team1_df.merge(
                        team2_df,
                        how="left",
                        on=["map_id", "map_name"],
                        suffixes=("_tm1", "_tm2") 
)
```

```python
# Create features with diffrence of values
# Lists with team features
diff_cols_list = team_df.drop(
                        ["map_name", "map_id", "team_id_tm1", "team_id_tm2"], 
                        axis=1
                        ).columns.to_list()


def make_diff_data(df, cols_list):

    """ Function creates features with differences of 2 teams features values
        Input:
                df[DataFrame]: dataframe with players' features
                cols_list[list]: list with names of features in df
        Return:
                DataFrame with features differences 
    """
    
    # Choose 2 columns from DataFrame which show same feature for 2 teams
    for i in range(len(cols_list)):
        for j in range(len(cols_list)):
            if cols_list[i].replace("_tm1","") == cols_list[j].replace("_tm2",""):
                # Create correct name of new feature
                col_name = cols_list[i].replace("_tm1","") + "_diff"
                # Create new feature with difference of values
                df[col_name] = df[cols_list[i]] - df[cols_list[j]]

    return df


df = make_diff_data(team_df, diff_cols_list)
df.drop(diff_cols_list, axis=1, inplace=True)
df
```

```python
# Encode categorical data
# Binary encoding
bin_encoder = ce.BinaryEncoder(cols=["team_id_tm1", "team_id_tm2"])
bin_enc = bin_encoder.fit_transform(df[["team_id_tm1", "team_id_tm2"]])
df = pd.concat([df, bin_enc], axis=1)

df.drop(["team_id_tm1", "team_id_tm2"], axis=1, inplace=True)

# OneHotEncoding 
df = pd.get_dummies(df, columns=["map_name"])
```

```python
train_df = train_df.merge(df, how="left", on=["map_id"])
test_df = test_df.merge(df, how="left", on=["map_id"])

drop_list = ["map_id", "map_name", "team1_id", "team2_id"]
train_df.drop(drop_list, axis=1, inplace=True)
test_df.drop(drop_list, axis=1, inplace=True)
```

```python
drop_cols = ["maps_played_diff", "rounds_played_diff",
             "kill_death_diff","rounds_with_kills_diff", "total_deaths_diff"]
train_df.drop(drop_cols, axis=1, inplace=True)
test_df.drop(drop_cols, axis=1, inplace=True)
```

```python
# Importance of features according to ANOVA test
X = train_df.drop(["who_win"], axis=1)
y = train_df["who_win"]

imp_feats = pd.Series(f_classif(X, y)[0], index=X.columns)
imp_feats.sort_values(inplace = True)

fig=px.bar(
    imp_feats,
    orientation="h",
    text_auto=" .5",
    height = 1200,
    width=1400,
    color=imp_feats.values,
    title="Importance of features according to ANOVA test",
    labels={"color":"Score range"}
)
fig.update_layout(
    font_family="Helvetica",
    xaxis=dict(tickfont_size=14, title="Score"),
    yaxis=dict(tickfont_size=12, title="Features")
)
fig.update_traces(
    textfont_size=12, 
    textangle=0, 
    textposition="outside", 
    cliponaxis=False
)
fig.show()
```

```python
# Drop columns with low score 
drop_cols = ["map_name_Nuke", "team_id_tm1_3", "map_name_Vertigo", 
             "saved_by_teammate_per_round_diff", "saved_teammates_per_round_diff"]
train_df.drop(drop_cols, axis=1, inplace=True)
test_df.drop(drop_cols, axis=1, inplace=True)
```

```python
# Split train dataset for train and validation parts
X = train_df.drop(["who_win"], axis=1)
y = train_df["who_win"]

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42, test_size=0.2)

# Check data
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_valid: {X_valid.shape}")
```

```python
# Use SelectKBest to select the 15 most relevant features
selector = SelectKBest(f_classif, k=10)
selector.fit(X_train, y_train)

# Create list of selected feature names 
list_of_features = list(selector.get_feature_names_out())
print(list_of_features)
```

```python
# Standardize data with StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(X_train[list_of_features])
X_train = pd.DataFrame(x_train_scaled, columns=list_of_features)
# Restore index for train data
X_train.set_index(y_train.index, inplace=True)

x_valid_scaled = scaler.transform(X_valid[list_of_features])
X_valid = pd.DataFrame(x_valid_scaled, columns=list_of_features)
# Restore index for test data
X_valid.set_index(y_valid.index, inplace=True)

test_index = test_df["index"]
test_df.drop("index", axis=1, inplace=True)
test_scaled = scaler.transform(test_df[list_of_features])
test_df = pd.DataFrame(test_scaled, columns=list_of_features)
# Restore index for test data
test_df.set_index(test_index, inplace=True)
```

```python
# Train model and calculate roc_auc metric
model_tree = tree.DecisionTreeClassifier(
    criterion="entropy",
    max_depth=5,
    min_samples_leaf=2,
    random_state=42
)
model_tree.fit(X_train, y_train)

y_pred_train = model_tree.predict(X_train)
y_pred_valid_prob = model_tree.predict_proba(X_valid)[:, 1]
tree_auc = metrics.roc_auc_score(y_valid, y_pred_valid_prob)

print(f"ROC AUC = {round(tree_auc, 3)}")
```

```python
submission = model_tree.predict(test_df)
submission = pd.DataFrame(columns=["who_win"], index=test_df.index, data=submission)

submission.to_csv("submission.csv")
```
